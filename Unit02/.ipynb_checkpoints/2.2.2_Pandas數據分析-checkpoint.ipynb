一個 DataFrame, 我們有 `index` (列的名稱), `columns` (行的名稱)。

## 4 增加一行

### 【技巧】

我們增加一行, 加入總級分。

```python
df['總級分'] = df[['國文', '英文', '數學', '自然', '社會']].sum(axis = 1)
df.head()
```

### 【技巧】

有計算的當然也可以的。

```python
df['加權總分'] = df.國文+df.英文+df.自然+df.社會+df.數學*2
print("加權後的資料：")
df.head()
```

## 5 排序和 index 重設

### 【重點】排序的方法

我們依總級分來排序。

```python
df.sort_values(by = "總級分",ascending = False).head(10) # 從大到小
```

### 【重點】排序的方法

加權分最高, 同分才看總級分

```python
dfSort = df.sort_values(by =["加權總分", "總級分"],ascending = False).head(10)
dfSort
```

### 【重點】重設 index

```python
dfSort = df.sort_values(by =["加權總分", "總級分"],ascending = False) # 先看加權總分再看總級分
dfSort.index =  range(1,101) # 重新排序
dfSort
```

## 6 篩出我們要的資料

基本上和 NumPy 的 array 篩法很像。

### 【重點】

找出數學滿級分同學。

```python
dfSort[dfSort.數學 == 15]
```

### 【重點】

找出數學和英文都滿級分的同學。要注意 `and` 要用 `&`, `or` 要用 `|`。每個條件一定要加弧號。

```python
dfSort[(dfSort.數學 == 15)&(dfSort.英文 == 15)]
```

## 7 刪除一行或一列

### 【重點】刪掉一行

我們來刪掉總級分的那行。

```python
dfSort.drop("總級分",axis = 1)
dfSort
```

### 【重點】改變原有的 DataFrame

我們會發現 `pandas` 很多動作都沒有更改原有的 DataFrame, 真的要改要加入

    inplace=True

```python
dfSort.drop("總級分",axis = 1, inplace=True)
dfSort
```

### 【重點】刪掉一列

刪掉列就是指定要刪去的 index。

```python
dfSort.drop(dfSort[dfSort.姓名 == "李士賢"].index) # inplace=True 先不設以免真被刪掉
dfSort
```

```python
dfSort.drop(dfSort[dfSort.姓名 == "李士賢"].index) # inplace=True 先不設以免真被刪掉
dfSort
```

### 【重點】刪掉一列

通常刪掉符合條件的比較合理 (注意是找到要刪掉的部份, 再找出相對的 index)。

```python
dfSort.drop(dfSort[dfSort.姓名 == "李士賢"].index) # inplace=True 先不設以免真被刪掉
dfSort
```

#### 20 日和 60 日的移動平均

```python
# 20日/60日的移動平均
df['Date'] = pd.to_datetime(df['Date'])  # 確保 Date 欄位是 datetime 格式
df.set_index('Date', inplace=True)       # 設定 Date 為索引
df300 = df[-300:]
df300.Close.plot(legend = True)
df300.Close.rolling(20).mean().plot(label = "$MA_{20}$",legend = True)
df300.Close.rolling(60).mean().plot(label = "$MA_{60}$",legend = True)

plt.show()
```

## 9 手工打造一個 DataFrame*

有時我們用手工打造一個簡單的 DataFrame, 可以更理解整個結構。其實很容易, 一個 DataFrame 基本上就包含兩個主要部份:

* 資料本身: 通常一個二維陣列 (矩陣)
* 行、列的名稱

我們來個簡單的小例子。

```python
mydata = np.random.randn(4,3)
```

```python
mydata = np.random.randn(4,3)
print(mydata)
```

把行列的名字放進去, 就成一個 DataFrame。我們列的部份先讓 Python 自己產生。

```python
df2 = pd.DataFrame(mydata, columns=list("ABC"))
```

```python
df2
```

#### 兩個表格上下貼起來

我們再來生一個 DataFrame, 再「貼」起來。

```python
df3 = pd.DataFrame(np.random.randn(3,3), columns=list("ABC"))
```

```python
df3
```

```python
df4 = pd.concat([df2,df3],axis = 0) # 因為是row要接(上下),若是Column要接則是 axis = 1
```

```python
df4
```

前面我們弄得亂七八糟的 index 重設一下。

```python
# 重設index
df4.index=range(7)
```

```python
df4
```

#### 横向的貼

```python
df5 = pd.concat([df2,df3],axis = 1)
```

等等, 這大小好像不太對也可以嗎? 答案是可以的!

```python
df5
```

#### 大一點的例子

我們來做前面「假的」學測資料。首先要有「假的」同學名單, 如果有興趣產生很多名字, 可以用這個服務。

[中文姓名產生器](http://www.richyli.com/name/index.asp)

```python
dfNames = pd.read_csv('data/names.csv',names = ["姓名"])
```

```python
dfNames.head()
```

```python
df_grades = pd.DataFrame(np.random.randint(6,16,(100,5)),
                        columns=['國文','英文','數學','社會','自然'])
```

```python
df_grades
```

```python
dfGrades2 = pd.concat([dfNames, df_grades], axis=1)  # 因為是row要接(上下),若是Column要接則是 axis = 1
print("新DataFrame:")
dfGrades2
```

```python
dfGrades2.to_csv("data/Grades2.csv",index = 0)
print("存檔至 'data/Grades2.csv'成功!")
```
