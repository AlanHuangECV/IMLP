```python
from sklearn.ensemble import BaggingClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=datasets.load_iris()
X=iris.data
y=iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=0)
```

```python
from sklearn import tree
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
```

```python
clf=tree.DecisionTreeClassifier()
# clf=svm.LinearSVC()
# clf=GaussianNB()
```

```python
# bagging=BaggingClassifier(base_estimator=clf,n_estimators=10,bootstrap=True,bootstrap_features=True,max_features=3,max_samples=0.7)
# base_estimator 已被棄用,改為estimator

"""
BaggingClassifier 參數說明：
estimator : 基礎估計器 (Base Estimator)。這是用來在每個隨機子集上訓練的模型。
如果為 None，則預設使用 DecisionTreeClassifier (決策樹分類器)。
n_estimators : 基礎估計器的數量。即套袋過程中將訓練多少個模型。
數值越大通常結果越穩定，但訓練時間也越長。
max_samples : 每個基礎估計器訓練時採用的樣本數。
可以是一個整數（表示具體的樣本數）或一個浮點數（表示佔總訓練集樣本數的比例）。
max_features : 每個基礎估計器訓練時採用的特徵數。
可以是一個整數（表示具體的特徵數）或一個浮點數（表示佔總特徵數的比例）。
bootstrap : 是否在採樣樣本時使用替換 (有放回採樣)。
如果為 True (預設值)，則使用有放回抽樣 (這就是標準的 Bagging)。如果為 False，則使用無放回抽樣。
bootstrap_features : 是否在採樣特徵時使用替換。如果為 True，則使用有放回抽樣來選擇特徵。
oob_score : 是否使用袋外 (Out-of-Bag, OOB) 樣本來評估泛化誤差。
如果為 True，則會計算那些未被用於訓練特定模型的樣本的預測分數。
n_jobs : 並行運行的 CPU 核心數。用於加速訓練。1 表示不使用並行，-1 表示使用所有可用的核心。
random_state : 隨機數種子。用於控制樣本和特徵採樣的隨機性，確保結果可重現。
verbose : 控制訓練時的冗長程度。用於在訓練過程中輸出進度或調試信息。
warm_start : 當設置為 True 時，重用先前呼叫的解決方案以適應更多的估計器；
否則，會清除並重新訓練所有估計器。
"""
bagging=BaggingClassifier(estimator =clf,n_estimators=10,bootstrap=True,bootstrap_features=True,max_features=3,max_samples=0.7)
```

```python
bagging.fit(X_train,y_train)
```

```python
bagging.predict(X_test)
```

```python
bagging.score(X_train,y_train)
```

```python
bagging.score(X_test,y_test)
```
